## What is Skin Cancer ?
 **Skin cancer is a type of cancer that develops in the skin. It occurs when skin cells are damaged, usually by exposure to ultraviolet (UV) radiation from the sun or tanning beds. There are several types of skin cancer, including basal cell carcinoma, squamous cell carcinoma, and melanoma.**

**Basal cell carcinoma is the most common type of skin cancer, accounting for about 80% of cases. It usually appears on areas of the skin that have been exposed to the sun, such as the face, neck, and arms. It often appears as a small, shiny, raised bump or a pink, scaly patch.**

**Squamous cell carcinoma is the second most common type of skin cancer, accounting for about 16% of cases. It also usually appears on areas of the skin that have been exposed to the sun, such as the face, neck, and arms. It often appears as a firm, red nodule or a flat, scaly lesion.**

**Melanoma is a less common but more dangerous type of skin cancer that can spread to other parts of the body if not detected and treated early. It often appears as a dark, irregularly shaped mole or spot on the skin.**

**It's important to get any suspicious moles or spots on your skin checked by a doctor, as early detection and treatment can greatly improve the chances of a full recovery.**

## What Is Skin Cancer Problems ?

**Skin cancer can cause a variety of problems, both physical and emotional. Here are some of the most common problems associated with skin cancer:** 

**1. Physical Discomfort: Skin cancer can cause physical discomfort in the form of pain, itching, and bleeding. As the cancer grows, it can become more painful and can even cause deformity in the affected area.** 

**2. Scarring: The surgical removal of skin cancer can leave scars, especially if the cancer is in a visible area like the face or neck. These scars can be disfiguring and may affect a person's self-esteem.**
    
 **3. Emotional Distress: A diagnosis of skin cancer can be very distressing and can cause anxiety, depression, and other emotional problems. Fear of recurrence can also be a significant source of stress for skin cancer survivors.**
    
**4. Financial Burden: The cost of skin cancer treatment can be high, especially if multiple surgeries are required. This can cause financial stress for some patients and their families.**
    
 **5. Increased Risk of Other Cancers: People who have had skin cancer are at higher risk of developing other types of cancer, including other skin cancers and internal cancers.**
    
**It's important to take steps to prevent skin cancer by protecting your skin from the sun, avoiding tanning beds, and regularly checking your skin for changes. If you notice any suspicious moles or spots, make an appointment with your doctor for evaluation.**

# **Building Deep Learning Model**

**This code defines a Convolutional Neural Network (CNN) using the Keras Sequential API to classify images of skin lesions as benign or malignant. Here's what each line does:**

**- `model = Sequential([...])` creates a new Keras Sequential model called `model`, which is initialized as an empty neural network.**

**- `data_augmentation` is a data augmentation pipeline defined in a previous code snippet.**

**- `Conv2D(16, (3,3), 1, activation="relu", padding="same")` creates a Convolutional layer with 16 filters, a kernel size of (3,3), a stride of 1, ReLU activation, and same padding. This layer convolves the input image with 16 different filters to extract features.**

**- `MaxPooling2D()` creates a MaxPooling layer that performs down-sampling along the spatial dimensions of the output feature map, by taking the maximum value of each non-overlapping rectangular patch.
**

**- The network architecture continues with more Conv2D and MaxPooling2D layers, with increasing number of filters, kernel sizes, and stride.**

**- `Flatten()` creates a Flatten layer that flattens the output of the last convolutional layer into a 1D array.**

**- `Dense(128, activation="relu")` creates a fully connected Dense layer with 128 units and ReLU activation. This layer receives the flattened output of the previous layer as input, and applies a linear transformation followed by a non-linear activation function to generate a feature vector.**

**- `Dense(1, activation="sigmoid")` creates a final Dense layer with 1 unit and sigmoid activation, which outputs a probability value between 0 and 1 indicating the predicted probability of the input image belonging to the positive class (i.e., malignant).**

**By stacking the different types of layers in the desired order, this model defines a CNN architecture that takes an input image, applies several convolutional and pooling layers to extract features, flattens the output into a 1D array, passes it through a fully connected layer to further process the features, and generates a binary prediction of the input image.**

```python
model = Sequential([
    data_augmentation,
    Conv2D(16, (3,3), 1, activation="relu", padding="same"),
    Conv2D(16, (3,3), 1, activation="relu", padding="same"),
    MaxPooling2D(),
    Conv2D(32, (5,5), 1, activation="relu", padding="same"),
    Conv2D(32, (5,5), 1, activation="relu", padding="same"),
    MaxPooling2D(),
    Conv2D(16, (3,3), 1, activation="relu", padding="same"),
    Conv2D(16, (3,3), 1, activation="relu", padding="same"),
    MaxPooling2D(),
    
    Flatten(),
    Dense(128, activation="relu"),
    Dense(1, activation="sigmoid")
])
```

**This code compiles a neural network model with the binary cross-entropy loss function, Adam optimizer, and three evaluation metrics: precision, recall, and accuracy.**

**- `loss='binary_crossentropy'`: This sets the loss function to binary cross-entropy, which is commonly used for binary classification problems.**

**- `optimizer='adam'`: This sets the optimizer to Adam, which is a popular optimizer used in deep learning.**

**- `metrics=[tf.keras.metrics.Precision(), tf.keras.metrics.Recall(), "acc"]`: This sets the evaluation metrics to be used during training and testing. Precision and recall are useful metrics for evaluating binary classification models, while "acc" represents the overall classification accuracy. The `tf.keras.metrics` module provides several pre-defined evaluation metrics that can be used directly in the `metrics` argument.**

```python
model.compile(loss='binary_crossentropy',
             optimizer='adam', 
             metrics=[tf.keras.metrics.Precision(), tf.keras.metrics.Recall(), "acc"])
```

During training, the model will update its weights and biases to minimize the loss function specified in the `compile` method. # **Results**

**This code evaluates the performance of the trained model on the test dataset using three metrics: binary accuracy, recall, and precision.** 

**First, `BinaryAccuracy()`, `Recall()`, and `Precision()` are initialized as objects to measure these metrics.** 

**Then, for each batch in the test dataset, the model predicts the output probabilities for the input images using `model.predict(X)`. These predicted probabilities `yhat` are then compared to the actual labels `y` to compute the metrics.** 
**
**`bin_acc.update_state(y, yhat)` updates the binary accuracy metric with the actual and predicted labels for the current batch. Similarly, `recall.update_state(y, yhat)` and `precision.update_state(y, yhat)` update the recall and precision metrics, respectively.** 

**Finally, the results of the metrics are printed using `bin_acc.result().numpy()`, `recall.result().numpy()`, and `precision.result().numpy()`. These methods return the current value of the metric, which is stored as a state in the corresponding metric object.**

```python
bin_acc = BinaryAccuracy()
recall = Recall()
precision = Precision()

for batch in test_data.as_numpy_iterator():
    X, y = batch
    yhat = model.predict(X)
    bin_acc.update_state(y, yhat)
    recall.update_state(y, yhat)
    precision.update_state(y, yhat)

print("Accuracy:", round(bin_acc.result().numpy(),2)," % ", "\nRecall:", recall.result().numpy(), "\nPrecision:", round(precision.result().numpy(),2)," % ")
```
